## Gold Mining Production Data EDA

```python
import pandas as pd
import numpy as np
```

```python
# Load the Excel file
file_path = 'Gold-Mining-Production-Volumes-Data.xlsx'  # Replace with your actual file path

# Load the Excel file, using the first row as header and skipping the initial title rows
df = pd.read_excel(file_path, sheet_name='Mine Production', skiprows=2)

df.head()
```

```python
df.loc[df['Country'] == 'DR Congo', 'Country'] = 'Democratic Republic of the Congo'
df.loc[df['Country'] == "Côte d'Ivoire", 'Country'] = 'Ivory Coast'
```

```python
# Save the dataframe as a CSV file
output_csv_path = 'Gold-Mining-Production-Volumes-Data.csv'

df.to_csv(output_csv_path, index=False)
```

```python
# - load idCountry.csv
df_countryid = pd.read_csv('idCountry.csv')
df_countryid.head()
```

```python
# Fill NaN values in CountryID based on partial string matching
df_merged = df.copy()
df_merged['CountryID'] = np.nan
for idx, row in df_merged[df_merged['CountryID'].isna()].iterrows():
    country_name = row['Country']
    match = df_countryid[df_countryid['name'].str.contains(country_name, case=False, na=False)]
    if not match.empty:
        df_merged.at[idx, 'CountryID'] = match['id'].values[0]
df_merged.head()
```
